```python
%matplotlib inline
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
t = np.arange(-1, 1, 50)
x = np.sin(t)
y = np.cos(t)

ax.plot3D(x, y, t, "o")
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)

plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
x = np.arange(start = -1, stop = 1, step = 0.02)
x_r = np.round(x, decimals = 3)
y = np.square(x)
y_r = np.round(y, decimals=3)
z = np.square(x)
z_r = np.round(z, decimals=3)

ax.plot3D(x_r, y_r, z_r, "o")
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
x = np.arange(start = -1, stop = 1, step = 0.02)
x_r = np.round(x, decimals = 3)
y = np.sin(x)
y_r = np.round(y, decimals=3)
z = np.cos(x)
z_r = np.round(z, decimals=3)

for i in range(len(x)):
    x[i] = 0

ax.plot3D(x, y_r, z_r, "o", color = "red")
print(x)
```
